```python
import collections
import math
import re
import string


def normalize_answer(s):
    """Lower text and remove punctuation, articles and extra whitespace."""

    def remove_articles(text):
        regex = re.compile(r"\b(a|an|the)\b", re.UNICODE)
        return re.sub(regex, " ", text)

    def white_space_fix(text):
        return " ".join(text.split())

    def remove_punc(text):
        exclude = set(string.punctuation)
        return "".join(ch for ch in text if ch not in exclude)

    def lower(text):
        return text.lower()

    #return white_space_fix(remove_articles(remove_punc(lower(s))))
    return (s)


def get_tokens(s):
    if not s:
        return []
    return normalize_answer(s).split()


def compute_f1(a_gold, a_pred):
    gold_toks = get_tokens(a_gold)
    pred_toks = get_tokens(a_pred)
    common = collections.Counter(gold_toks) & collections.Counter(pred_toks)
    num_same = sum(common.values())
    if len(gold_toks) == 0 or len(pred_toks) == 0:
        # If either is no-answer, then F1 is 1 if they agree, 0 otherwise
        return int(gold_toks == pred_toks)
    if num_same == 0:
        return 0
    precision = 1.0 * num_same / len(pred_toks)
    recall = 1.0 * num_same / len(gold_toks)
    f1 = (2 * precision * recall) / (precision + recall)
    return f1
```

```python
def remove_HTML_syntax(input_text):
    text = ''
    accumulate = False
    output_text = input_text
    for char in input_text:
        if char == '<':
            accumulate = True
        elif char == '>':
            output_text = output_text.replace(text + char, '')
            text = ''
            accumulate = False
        
        if accumulate:
            text += char


    return output_text
```

```python
import re
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
def getSCGArticle(article_name):
  with open('/content/wiki-documents-nsc/' + article_name, encoding="utf-8") as f:
    lines = f.readlines()
    raw = ""
    hashtag= ""
    for line in lines:
      if not line.startswith("#") and len(line.strip()) >= 2:
        line = deEmojify(line)
        line = line.replace("”", "").replace("“", "").replace("\"", "").replace("\ufeff", "").replace("\u200b", "")
        line = line.strip()
        raw += line + "\n"
      elif line.startswith("#"):
        hashtag += line.strip()
  return raw
getSCGArticle('SCG101.txt') 
```

```python
def getArticle (article_name):
  if ('SCG' in article_name):
    return getSCGArticle(article_name)
  try:
    with open('/content/wiki-documents-nsc/' + article_name, encoding="utf-8") as f:
      lines = f.readlines()
      lines  = remove_HTML_syntax(lines[0])
    return lines
  except:
    return ''

getArticle('SCG217.txt')

```

```python
import pandas as pd
import numpy as np
import os
from tqdm.notebook import tqdm
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
from difflib import SequenceMatcher
import re
import torch
```

```python
#pretrained = "deepset/xlm-roberta-large-squad2"
#pretrained = "cstorm125/wangchanberta-base-att-spm-uncased-finetune-qa"
pretrained = "cstorm125/xlm-roberta-base-finetune-qa"
#pretrained = "cstorm125/bert-base-multilingual-cased-finetune-qa"
#pretrained = "cstorm125/wangchanberta-base-wiki-20210520-news-spm_span-mask-finetune-qa"
tokenizer = AutoTokenizer.from_pretrained(pretrained ,  max_length=512)
model = AutoModelForQuestionAnswering.from_pretrained(pretrained)
```

```python
if torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')
model.to(device);
```

```python
def find_answer(question, context, tokenizer=tokenizer, model=model, device=device):
    def predict(inputs, model):
        input_ids = inputs["input_ids"].tolist()[0]
        text_tokens = tokenizer.convert_ids_to_tokens(input_ids)
        answer_start_scores, answer_end_scores = model(**inputs)
        answer_start = torch.argmax(answer_start_scores)  
        answer_score = torch.max(answer_start_scores) +  torch.max(answer_end_scores)
        answer_end = torch.argmax(answer_end_scores) 
        answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))
        preanswer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start-1:answer_start]))

        # output = model(**inputs)
        # answer_start_scores = output['start_logits']
        # answer_end_scores = output['end_logits']
        # answer_start = torch.argmax(answer_start_scores)  
        # answer_score = torch.max(answer_start_scores)
        # answer_end = torch.argmax(answer_end_scores) + 1
        # answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))
        return answer, answer_score , preanswer
    
    inputs = tokenizer(question, context, add_special_tokens=True, return_tensors="pt" ).to(device)
    input_size = inputs.input_ids.size()[1]
    answer, answer_score  ,preanswer = "<s>", -1 ,'<s>'
    if input_size > 512:
        max_length = 300
        chunks = [context[i:i+max_length] for i in range(0, len(context), max_length)]
        max_chunk_answer_score = -1
        for chunk in chunks:
            inputs = tokenizer(question, chunk, add_special_tokens=True, return_tensors="pt", max_length=512, truncation=True).to(device)
            chunk_answer, chunk_answer_score , chunk_preanswer = predict(inputs, model)
            num_chunk_answer_score = chunk_answer_score.detach().cpu().numpy()
            if chunk_answer != "<s>" and num_chunk_answer_score > max_chunk_answer_score:
                answer, answer_score ,preanswer = chunk_answer, chunk_answer_score , chunk_preanswer
                max_chunk_answer_score = num_chunk_answer_score
    else:
        answer, answer_score , preanswer = predict(inputs, model)
    return answer, answer_score , preanswer
```

```python
def deletePep(text):
  if text!=text:
    return None
  regex = re.compile(r"^ใน", re.UNICODE)
  text  = re.sub(regex, "", text)
  regex = re.compile(r"^ที่", re.UNICODE)
  text =  re.sub(regex, "", text)
  regex = re.compile(r"^ของ", re.UNICODE)
  text =  re.sub(regex, "", text)
  regex = re.compile(r"^สี", re.UNICODE)
  text =  re.sub(regex, "", text)
  regex = re.compile(r"^เป็น", re.UNICODE)
  text =  re.sub(regex, "", text)
  regex = re.compile(r"^และ", re.UNICODE)
  text =  re.sub(regex, "", text)
  regex = re.compile(r"^##", re.UNICODE)
  text =  re.sub(regex, "", text)
  regex = re.compile(r"^\'", re.UNICODE)
  text =  re.sub(regex, "", text)
  return text
deletePep('ที่หิมะในทุ่มดอกไม้ABCD')
```
