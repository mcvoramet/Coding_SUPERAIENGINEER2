```python
import onnx
from onnxruntime.quantization import quantize, QuantizationMode
```

```python
# Load the onnx model
model = onnx.load('/content/drive/MyDrive/POSTAG/BERT_model/ONNXModel(Noquantize)/onnx_model.onnx')
# Quantize
quantized_model = quantize(model, quantization_mode=QuantizationMode.IntegerOps)
# Save the quantized model
onnx.save(quantized_model, '/content/drive/MyDrive/POSTAG/BERT_model/QuantizeTest/quantized_model.onnx')
```
