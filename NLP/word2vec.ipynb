```python
# TOO SLOW
def get_text(data, idx):
  out = []
  for i in range(len(data[idx]['char'])):
    if data[idx]['is_beginning'][i]:
      out.append(' ')

    out.append(data[idx]['char'][i])

  return ''.join(out)
```

```python
''.join(dataset['train'][0]['char'])
```

```python
get_text(dataset['train'], 0)
```

```python
sentences = []
for i in range(100):
  sentences.append(get_text(dataset['train'], i))
```

```python
import numpy as np
```

```python
idx = []
sentences = []
for tt in range(20000):
  t = np.random.randint(len(dataset['train']))
  idx.append(t)

  o = list(dataset['train'][t]['char'])

  for i,j in enumerate(np.where(np.array(dataset['train'][t]['is_beginning']) == 1)[0]):
    o.insert(i+j, ' ')

  sentences.append(''.join(o))
```

```python
sentences[10]
```

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence

#sentences = LineSentence('txt.txt')

model = Word2Vec([s.split() for s in sentences], 
                 size=500, window=10, min_count=10, iter=100, workers=4)
```

```python
model
```

```python
vocab = list(model.wv.vocab.keys())
```

```python
def analogy( model, pos, neg ):
    l = model.wv.most_similar(positive=pos, negative=neg)
    s,t = l[0]
    print(s)
```
